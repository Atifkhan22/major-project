```python
genes = ["FAS", "FADD", "TNFRSF10B", "CASP8", "CASP10", "BAX",	"BAK1", "BCL2", "BCL2L1", "BID", "BAD", "CYCS", "APAF1", "CASP9", "CASP3", "CASP6", "CASP7", "XIAP", "DIABLO", "TP53", "MDM2", "ATM", "ATR", "CDKN1A", "AKT1", "PIK3CA","MAPK1", "MAPK3",	"NFKB1", "RELA",] 


folder_path = "C:/Users/aatif/Downloads/ml part in project/LUAD_TUMOR_RAW"

data_list = []

for file in os.listdir(folder_path):
    if file.endswith(".tsv"):
        file_path = os.path.join(folder_path, file)
        
        df = pd.read_csv(file_path, sep="\t", comment="#")
        
        # Keep only needed columns
        df = df[["gene_name", "tpm_unstranded"]]
        
        # Clean unwanted rows
        df = df[df["gene_name"].notna()]
        df = df[~df["gene_name"].str.startswith("N_")]
        
        # Keep only your genes
        df = df[df["gene_name"].isin(genes)]
        
        # Convert to dictionary
        gene_dict = dict(zip(df["gene_name"], df["tpm_unstranded"]))
        
        # Create row (ensure all genes exist)
        row = [gene_dict.get(gene, 0) for gene in genes]
        
        # Add label (1 = tumor)
        row.append(1)
        
        data_list.append(row)

# Create DataFrame
columns = genes + ["Label"]
tumor_df = pd.DataFrame(data_list, columns=columns)
```

```python
normal_path = "C:/Users/aatif/Downloads/ml part in project/LUAD_NORMAL_RAW"

data_list = []   # VERY IMPORTANT (reset here)

for file in os.listdir(normal_path):
    if file.endswith(".tsv"):
        file_path = os.path.join(normal_path, file)
        
        df = pd.read_csv(file_path, sep="\t", comment="#")
        
        # Keep required columns
        df = df[["gene_name", "tpm_unstranded"]]
        
        # Clean unwanted rows
        df = df[df["gene_name"].notna()]
        df = df[~df["gene_name"].str.startswith("N_")]
        
        # Keep only selected genes
        df = df[df["gene_name"].isin(genes)]
        
        # Convert to dictionary
        gene_dict = dict(zip(df["gene_name"], df["tpm_unstranded"]))
        
        # Create row (ensure all genes present)
        row = [gene_dict.get(gene, 0) for gene in genes]
        
        # Add label (0 = normal)
        row.append(0)
        
        data_list.append(row)

# Create DataFrame
columns = genes + ["Label"]
normal_df = pd.DataFrame(data_list, columns=columns)
```
