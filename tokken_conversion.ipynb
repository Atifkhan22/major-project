```python
import pandas as pd

df = pd.read_csv("Final_Dataset_Human_Chimp.csv")

df.head()
```

```python
import numpy as np

df["Norm_Normal"] = df["Mean_TPM_Normal"] / df["Mean_TPM_Normal"].max()
df["Norm_Tumor"] = df["Mean_TPM_Tumor"] / df["Mean_TPM_Tumor"].max()
df["Norm_Chimp"] = df["Chimp_Expression"] / df["Chimp_Expression"].max()
```

```python
def assign_tokens(x):
    if x <= 0.15:
        return 1
    elif x <= 0.3:
        return 2
    elif x <= 0.5:
        return 3
    elif x <= 0.7:
        return 4
    elif x <= 0.85:
        return 5
    else:
        return 6
```

```python
df["Tokens_Normal"] = df["Norm_Normal"].apply(assign_tokens)
df["Tokens_Tumor"] = df["Norm_Tumor"].apply(assign_tokens)
df["Tokens_Chimp"] = df["Norm_Chimp"].apply(assign_tokens)
```

```python
df[["gene_name","Tokens_Normal","Tokens_Tumor","Tokens_Chimp"]]
```

```python
df["Norm_Chimp"] = (df["Chimp_Expression"] - df["Chimp_Expression"].min()) / (df["Chimp_Expression"].max() - df["Chimp_Expression"].min())
```

```python
df["Tokens_Chimp"] = df["Norm_Chimp"].apply(assign_tokens)
```

```python
df[["gene_name","Tokens_Chimp"]]
```

```python
final_tokens = df[["gene_name", "Tokens_Normal", "Tokens_Tumor", "Tokens_Chimp"]]
```

```python
final_tokens = final_tokens.sort_values(by="gene_name")
```

```python
final_tokens = final_tokens.reset_index(drop=True)
```

```python
final_tokens.to_csv("Final_Tokens_Dataset.csv", index=False)
```
